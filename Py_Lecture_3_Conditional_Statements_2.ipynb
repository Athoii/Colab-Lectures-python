Let's consider you have a fashion house. For Eid, you are providing discount to the customers based on the purchased amount. Your policy is as follows, if a customer purchase over 10k then she will get 15% discount otherwise, 7.5% discount.

Write a python program that will do the following:


1.   Take the purchase amount as input from the user (either, integer or float)
2.   Using conditional statement calculate the disount amount (as per above policy) and the payable amount.
3. Print the discount amount and payable amount.



```python
purchase_amount=int(input('enter your purchase amount: '))
discount=0
if(purchase_amount>10000):
  discount=(purchase_amount*15)/100
else:
  discount=(purchase_amount*7.5)/100

print('Discount: ',discount)
print('Payable Amount',(purchase_amount-discount))
```

#**IF...ELIF...ELSE Statement**

if-else-if.png

**If you have multiple (more than two) mutually exclusive conditions to check and decide, then you have to use,**

                                     IF...ELIF...ELSE

**Example**
Let's consider you have a fashion house. For Eid, you are providing discount to the customers based on the purchase amount. Your policy is as follows,
if a customer purchase over 50k then she will get 22% discount.

1.   if the purchase amount is above 50k then 22% discount.
2.   if the purchase amount is between 40k and 49999 then she will get 20% discount.
3.   if the purchase amount is between 25k and 39999 then she will get 10% discount.
4.   Otherwise, there is no discount.

Write a python program that will perform the followings:
1.   Take the purchage amount as input from the user (either, integer or float)
2.   Using conditional statement calculate the disount amount (as per above policy) and the payable amount.
3. Print the discount amount and payable amount.

```python
amount = int(input('Enter purchage amount: '))


if amount > 50000:
  discount = amount * .22
elif amount > 40000 and amount < 49999:
  discount = amount * .20
elif amount > 25000 and amount < 39999:
  discount = amount * .10
else:
  discount = 0

payable_amount = amount - discount

print(f'Purchase amount: {amount} bdt')
print(f'Discount amount: {discount} bdt')
print(f'Payable amount: {payable_amount} bdt')
```
