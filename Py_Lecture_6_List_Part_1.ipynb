# **Adding elements in the List**

**Add / Insert value with append() function.**


---



```python
numbers = list(range(11,20))

print('number list: ', numbers)
print('no of items: ', len(numbers))

print()
# insert a value 999 to the list
numbers.append(999)
print('after add the value: ', numbers)
print('no of items: ', len(numbers))
```

**Add / Insert multiple values with extend() function.**


---

```python
numbers = list(range(11,20))

print('number list: ', numbers)
print('no of items: ', len(numbers))

# insert all these values 777, 420, 840

numbers.extend([777, 420, 840])
print('after adding all the values: ', numbers)
print('no of items: ', len(numbers))
```

**Add / Insert a value at a specific location in the list.**


---

```python
myList = ['Bangladesh', 170, 'India', 'Pakistan', 20.75, 'apple', 'pen', 1998, 2005761.98]

print('Before updating a value: ', myList)
print('No of items: ', len(myList))

print()
# insert West Indies at location 2 in the list
myList.insert(2, 'West Indies')
print('After inserting value at 2nd place: ', myList)
print('No of items: ', len(myList))
```

# **Removing elements from the List**

**Remove with del command**


---



**Use the del statement if you know exactly which element(s) you are deleting**

```python
listOfDiff = ['Bangladesh', 170, 'India', 'Pakistan', 20.75, 'apple', 'pen', 1998, 2005761.98]

print('Before delete: ', listOfDiff)
print('No of items: ', len(listOfDiff))

print()
#delete 4th element from the list
del listOfDiff[3]
print('After deleting 4th element delete: ', listOfDiff)
print('No of items: ', len(listOfDiff))


print()
#delete last three items from the list
del listOfDiff[5:8]
print('After deleting last three items: ', listOfDiff)
print('No of items: ', len(listOfDiff))

#delete entire list
del listOfDiff
print(listOfDiff)
```

**Remove with remove(obj) function**


---

**Remove a specific element from the list.**

```python
myList = [100, 50, 21.23, 'sazzad', '100', 3, 6, 101, 100]

print('Before delete: ', myList)
print('No of items: ', len(myList))

print()
#delete 100 from the list
myList.remove(100)
#myList.remove(100)

print('After deleting 100 from the list: ', myList)
print('No of items: ', len(myList))
```

**Remove with pop(index) function**


---

**Remove and get the last element or a a specific element (identified by the index) from the list.**

```python
myList = [100, 50, 21.23, 'sazzad', '100', 3, 6, 101, 'i am the last']

print('Before pop: ', myList)
print('No of items: ', len(myList))

print()
# pop the last element from the list
last = myList.pop()
print('After popping the last element: ', myList)
print('No of items: ', len(myList))
print('The popped element: ', last)

print()
# pop the last element from the list
last = myList.pop()
print('After popping the last element: ', myList)
print('No of items: ', len(myList))
print('The popped element: ', last)

print()
# pop the 4th element from the list
myItem = myList.pop(3)
print('After popping the 4th element: ', myList)
print('No of items: ', len(myList))
print('The popped element: ', myItem)
```

# **Other useful functions for List manipulation**

**count(obj) functions counts how many times a given value appears in the list**

---



```python
myList = [100, 50, 21.23, 'sazzad', '100', 3, 6, 101, 100, 'i am the last', 50, 50]

x = myList.count(100)
print(f'100 appears {x} times in the list')

y = myList.count(50)
print(f'50 appears {y} times in the list')
```

**.index() method gives the exact index of a particular value in the list**

```python
myList = [100, 50, 21.23, 'sazzad', '100', 3, 6, 101, 100, 'i am the last', 50, 50]

# find the index for the value 'sazzad'
x = myList.index('sazzad')
print(f'sazzad is at location {x}')
```

---

```python
myList = [100, 50, 21.23, 'sazzad', '100', 3, 6, 101, 100, 'i am the last', 50, 50]

print('The original list: ', myList)

# reversing the entire list
myList.reverse()
print('The reverse list: ', myList)
```
