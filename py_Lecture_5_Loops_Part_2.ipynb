#**Break statement in Loop**

image.png

**The break statement in Python terminates the current loop and resumes execution at the next statement, just like the traditional break found in C.**

**Example 1: Break Statement**

```python
# loop without a break statement
i = 1
while i < 6:
   print('i = ', i)
   i+=1
print('The program ends.')
```

```python
# Loop with a break statement
i = 1
while i < 6:
   print('i = ', i)
   i+=1
   break

print('The program ends.')
```

**Example 2: break statement with a condition checking**

```python
i = 1
while i < 6:
   print('i = ', i)
   i+=1
   if i == 3:
      print('program breaks out from here!!')
      break
print('The program ends.')
```

#**Nested Loops**

**Python programming language allows the use of one loop inside another loop. The following section shows a few examples to illustrate the concept.**

**Example: 1 Write a python program using loop to print a square box of * for a given number**

Enter a number: 4

\*  *  *  *

\*  *  *  *

\*  *  *  *

\*  *  *  *

```python
N = int(input('Enter a number:'))#5

i = 1
while i <= N:
  j = 1
  while j <= N:
    print('*', end = ' ')
    j +=1
  print()
  i +=1
```

# **The range() function**

**This function can be used to iterate over a range of values in a sequence (e.g., list, tuples) of values.**

**Syntax:**
  
  range(stop) --> end value, but end value is excluded, start will be by default 0, step by default +1

  range(start, stop) --> step by default +1
  
  range(start, stop, step) --> step = increment/decrement

```python
print(range(10))

for i in range(10):
  print(i, end = ', ')
```

```python
print(range(4,10))

for i in range(4,10):
  print(i, end = ', ')
```

```python
print(range(50, 10, -2))

for i in range(50, 10, -2):
  print(i, end = ', ')
```
