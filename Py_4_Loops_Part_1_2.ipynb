**Example: 2** Write a Python Program to find the sum of series 1^2+2^2+3^2...+N^2.

```python
N = int(input('Enter End Number: '))

number = 1
sum = 0
while number <= N:
  print(f'{number}^2', end ="+")
  sum += pow(number,2)
  number += 1

print(f'={sum}')
```

**Exercise: 4** Write a Python Program to find the sum of series 1^1+2^2+3^3...+N^N

```python
N=int(input('enter end number'))
number=1
sum=0
while number <=N:
  print(f'{number}^{number}',end ='+')
  sum += pow(number,N)
  number += 1

print(f' ={sum}')
```

**Exercise: 5** Write a Python program to display the n terms of odd natural number and their sum.

Sample Output:

Enter end number: 10

1 + 3 + 5 + 7 + 9 + = 25

```python
N = int(input("Enter end number: "))
start_Number = 1
current_Number = start_Number
total_Sum = 0

while current_Number <= N:
    if current_Number % 2 == 1:
        total_Sum += current_Number
        print(f'{current_Number}', end=" + ")
    current_Number += 1

print(f' = {total_Sum}')
```

**Exercise: 6** Write a Python program to display the n terms of even natural number and their sum.

Sample Output:

Enter end number: 10

2 + 4 + 6 + 8 + 10 + = 30

```python
N = int(input("Enter end number: "))
start_Number = 2
current_Number = start_Number
total_Sum = 0

while current_Number <= N:
    if current_Number % 2 == 0:
        total_Sum += current_Number
        print(f'{current_Number}', end=" + ")
    current_Number += 1

print(f' = {total_Sum}')
```

**Exercise: 6** Write a Python program to calculate the factorial of a given number.

```python
fact_number=int(input("Enter the number: "))
result=1
for i in range (1,fact_number+1):
  result*=i

print(result)
```

**Exercise: 7** Write a Python Program to read the CGPAs of 5 students and print the sum of the CGPAs and average CGPA.

```python
count=1
sum=0
avg=0
while(count<=5):
  cgpa=float(input('Enter the cgpa'))
  sum+=cgpa
  count+=1
avg=sum/5
print('Sum:',sum)
print('Average:',avg)
```

**Example: 4** Write a Python Program to read the CGPAs of 5 students and print the max CGPA, average CGPA.

```python
max_gpa = 0
min_gpa = 4 # considering the max gpa is 4

count  = 1
while count <= 5:
  gpa = float(input('Enter GPA: '))#3.5 , 2.7, 3.99

  if max_gpa < gpa: #0<3.5
    max_gpa = gpa #3.5,3.99

  if min_gpa > gpa: #4>3.5
    min_gpa = gpa #3.5,2.7

  count += 1

print(f'Max Gpa is: {max_gpa}')
print(f'Min Gpa is: {min_gpa}')
```

**Example 5:** Write a Python program to display the cube of the number upto given an integer.

```python
N = int(input('Upto which number: '))

number = 1

while(number <= N):
  cube = pow(number,3)
  print(f'Cube of number {number} is {cube}')
  number += 1

print('end of program.. byee')
```

**Exercise: 8** Write a Python program to display the multiplication table of a given integer.
